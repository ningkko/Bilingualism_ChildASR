```python
import pandas as pd
import numpy as np
```

```python
def unify(df):
    df.columns = ["child_ID","age"]+[i.replace("\n","").replace(" ","") for i in list(df.columns)[2:]]
    non_repetitive_column = []
    for i in df.columns:
        if i in non_repetitive_column:
            non_repetitive_column.append(i+"_2")
        else:
            non_repetitive_column.append(i)
    df.columns = non_repetitive_column
    df = df.loc[:,~df.columns.str.contains('\.')].set_index("child_ID")
    return df
```

```python
spanish_1 = unify(pd.read_csv("spanish/paper/by_child_data.csv"))
```

```python
spanish_2 = unify(pd.read_csv("Spanish_Marchman/paper/by_child_data.csv"))
```

```python
irish = unify(pd.read_csv("irish/paper_data/by_child_data.csv"))
```

```python
hebrew = unify(pd.read_csv("hebrew/paper_data/by_child_data.csv"))
```

```python
french = unify(pd.read_csv("french/paper_data/by_child_data.csv"))
```
