## process all data seperately

```python
import pandas as pd
```

```python
spanish_hdpm = pd.read_csv("spanish/paper/by_word_by_age(HDPM).csv")
spanish_r01 = pd.read_csv("spanish/paper/by_word_by_age(R01).csv")
spanish = pd.merge(spanish_hdpm, spanish_r01,on="Unnamed: 0", how="inner")
spanish
```

```python
spanish = spanish[1:]
item = [i.replace(".0","").replace("\n","").replace(" ","") for i in spanish.pop("Unnamed: 0")]
spanish.insert(0, "item", item)
spanish.columns = [i.replace(".0","") for i in list(spanish.columns)]
spanish = spanish.set_index(spanish.item)
spanish = spanish.transpose()[1:].transpose()
spanish = spanish.astype(float).drop_duplicates()
spanish
```

```python
irish = pd.read_csv("irish/paper_data/english_in_irish.csv")

irish = irish[2:627]
item = [i.replace(".0","").replace("\n","").replace(" ","")  for i in irish.pop("Unnamed: 0")]
irish.insert(0, "item", item)
irish = irish.set_index(irish.item)
irish = irish.transpose()[1:].transpose()
irish = irish.astype(float).drop_duplicates()
irish
```

```python
hebrew = pd.read_csv("hebrew/paper_data/by_word_by_month.csv")
hebrew
```

```python
hebrew = hebrew[1:487]
item = [i.replace(" ","").replace("\n","").replace(" ","")  for i in hebrew.pop("Unnamed: 0")]
hebrew.insert(0, "item", item)
hebrew = hebrew.set_index(hebrew.item)
hebrew = hebrew.transpose()[1:].transpose()
hebrew = hebrew.astype(float).drop_duplicates()
hebrew
```

```python
french_wave2 = pd.read_csv("french/paper_data/by_word_by_age(wave2).csv")
french_wave3 = pd.read_csv("french/paper_data/by_word_by_age(wave3).csv")
french = pd.concat([french_wave2.transpose(),french_wave3.transpose()])
french = french.transpose()[2:].transpose()[1:].transpose()
french
```

```python
item = [i.replace(" ","").replace("\n","").replace(" ","")  for i in french.pop("Unnamed: 0")]
french.insert(0, "item", item)
french = french.set_index(french.item)
french = french.transpose()[1:].transpose().astype(float).drop_duplicates()
french
```

## Marchman Spanish data

```python
spanish_marchman = pd.read_csv("Spanish_Marchman/paper/by_word_by_age.csv")
spanish_marchman = spanish_marchman[1:]
item = [i.replace(".0","").replace("\n","").replace(" ","") for i in spanish_marchman.pop("Unnamed: 0")]
spanish_marchman.insert(0, "item", item)
spanish_marchman = spanish_marchman.set_index(spanish_marchman.item)
spanish_marchman = spanish_marchman.transpose()[1:].transpose()
spanish_marchman = spanish_marchman.astype(float).drop_duplicates()
spanish_marchman
```

```python
def add_lang(lang,df):
    df["lang"] = [lang]*len(df)
    
    return df.reset_index()

french = add_lang("french",french)
hebrew = add_lang("hebrew",hebrew)
spanish = add_lang("spanish",spanish)
irish = add_lang("irish",irish)
spanish_marchman = add_lang("spanish",spanish_marchman)
```

```python
hebrew
```

```python
import string
df_all = pd.concat([hebrew,irish,spanish,french,spanish_marchman],ignore_index=True)
df_all = df_all.reindex(sorted(df_all.columns), axis=1).dropna(how="all")
df_all.item = df_all.item.apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
df_all
```

```python
import numpy as np
df_mean = df_all.groupby("item").agg(np.mean)
df_mean = df_mean.sort_index(key=lambda x: x.str.lower())
df_mean
```
