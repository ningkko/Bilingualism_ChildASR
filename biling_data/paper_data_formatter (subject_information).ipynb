## Irish - English
https://www-tandfonline-com.ezp-prod1.hul.harvard.edu/doi/pdf/10.1080/13670050.2016.1179258?needAccess=true


```python
irish = pd.read_csv("irish/paper_data/english_in_irish.csv")
irish_subject = pd.read_csv("irish/paper_data/individual_info.csv")

## country 
# The participants in this group were living in Irish-speaking communities in the
# west of Ireland, but most were also exposed to English from birth.
irish_subject.insert(1,"Country",["Ireland"]*99)
irish_subject
```

```python
# convert MainCaregiverIrish and MainCaregiverEnglish into 5-point scales
# proportionEnglish1 = MainCaregiverEnglish / (MainCaregiverEnglish + MainCaregiverIrish)
irish_subject.MainCaregiverIrish.unique()
```

```python
def map_irish_maincare(x):
    if x=='Always':
        return 5
    elif x=='Usually':
        return 4
    elif x=="50:50":
        return 3
    elif x=='sometimes':
        return 2
    elif x =='Never/rarely':
        return 1 
    else:
        return np.nan
    
irish_subject.MainCaregiverIrish = irish_subject.MainCaregiverIrish.apply(lambda x: map_irish_maincare(x))
irish_subject.MainCaregiverEnglish = irish_subject.MainCaregiverEnglish.apply(lambda x: map_irish_maincare(x))
    
```

```python
irish_subject
```

```python
irish_subject.LanguageHousehold.unique()
```

```python
# convert LanguageHousehold to a scale: EnglishOnly = 1, Mostly English = .75, Equal Irish & ENglish = .5, Mostly Irish = .25, Irish only = 0
# Average proportionEnglish1 and LanguageHousehold

def map_irish_household(x):
    if x =='Mostly Irish':
        return 0.25
    elif x =='Equal Irish and English':
        return 0.5
    elif x=='Mostly English':
        return 0.75
    elif x=='Irish Only':
        return 0
    elif x=='English Only':
        return 1
    else:
        return np.nan
    
irish_subject.LanguageHousehold = irish_subject.LanguageHousehold.map(lambda x: map_irish_household(x))
irish_subject

```

```python
## Average proportionEnglish1 (converted to scale 0-1) and LanguageHousehold
irish_subject["Proportion of time"] = ((irish_subject.MainCaregiverIrish/(irish_subject.MainCaregiverIrish+irish_subject.MainCaregiverEnglish))/5+irish_subject.LanguageHousehold)/2
irish_subject
```

## English - Hebrew
https://www-tandfonline-com.ezp-prod1.hul.harvard.edu/doi/pdf/10.1080/13670050.2016.1179257?needAccess=true

The present study explores the vocabulary development of bilingual
children when neither of their languages has a minority language status.
With both languages having high relative prestige, it is possible to
address the impact of exposure variables: age of onset, length of 
exposure, and frequency of exposure (FoE) to both languages. 


```python
hebrew = pd.read_csv("hebrew/paper_data/by_word_by_month.csv")
hebrew_subject = pd.read_csv("hebrew/paper_data/individual_info.csv")[:40]
def m_gender(x):
    if x==1:
        return "Female"
    elif x==2:
        return "Male"
    else:
        return ""

hebrew_subject.gender = hebrew_subject.gender.map(lambda x:m_gender(x))
session = hebrew_subject.pop("ageband")
hebrew_subject.insert(1,"session",session)
hebrew_subject
```

```python
hebrew_subject.columns = [i.replace("L1","English").replace("L2","Hebrew") for i in list(hebrew_subject.columns)]
hebrew_subject.session = hebrew_subject.session.astype(int)
hebrew_subject["Age of Hebrew exposure"] = hebrew_subject["Age of Hebrew exposure"].astype(int)
hebrew_subject
```

```python
## Hebrew: proportionL1 = Freqexposurel1 / (Freqexposurel1 + Freqexposurel2)
### converted 50 1-5 scale, normalized 
hebrew_subject["Proportion of time"] = ((hebrew_subject.FreqexposureEnglish+1)/5+(hebrew_subject.FreaexposureHebrew+1)/5)/2   
hebrew_subject = hebrew_subject.drop(["FreqexposureEnglish","FreaexposureHebrew"], axis=1)
hebrew_subject.insert(1,"Country",["Israel"]*40)
hebrew_subject
```

```python
hebrew_subject.to_csv("hebrew_info.csv", index=False)
```
