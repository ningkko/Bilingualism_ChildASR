```python
items = []
section = []
lang = []
gloss = []
for index, row in df.iterrows():
#     print(row)
    if row["c0"]:
        section_name = row["c1"]
    else:
        # in the section, look for items in english and potential items in hebrew
        # if universal, out the same word for 2
        if row["c1"]:
            items.append(row["c1"])
            if not row["c2"]:
                gloss.append(row["c1"])
                lang.append("universal")
            else:
                gloss.append(row["c1"])
                lang.append("Eng")
                
            section.append(section_name)

            if row["c2"]:
                items.append(row["c2"])
                lang.append("Heb")
                section.append(section_name)
                gloss.append(row["c1"])


            if row["c3"]:
                items.append(row["c3"])
                lang.append("Heb")
                section.append(section_name)
                gloss.append(row["c1"])

            if row["c4"]:
                items.append(row["c4"])
                lang.append("Heb")
                section.append(section_name)
                gloss.append(row["c1"])




        
```

```python
import numpy as np
instruction = pd.DataFrame(np.array([items, section, items, gloss, lang]).transpose(), columns = ["item","type","definition","gloss","language"])
instruction
```

```python
choices = ["comprehension/production"]*len(instruction)
```

```python
instruction["choices"] = choices
```

```python
instruction["itemID"] = ["item_"+str(i) for i in list(range(1,len(instruction["item"])+1))]
instruction["complexity_category"] = ['']*len(instruction["item"])
```

```python
from collections import Counter
items = list(instruction.item)
uniq = []
for i in items:
    if i not in uniq:
        uniq.append(i)
    else:
        uniq.append(i+"_2")
instruction.item = uniq
```

```python
from xlrd import open_workbook
import webcolors

wb = open_workbook('English-Hebrew data for WordBank.xls', formatting_info=True)
sh = wb.sheet_by_name('Child1')

def getBGColor(book, sheet, row, col):
    xfx = sheet.cell_xf_index(row, col)
    xf = book.xf_list[xfx]
    bgx = xf.background.pattern_colour_index
    pattern_colour = book.colour_map[bgx]

    #Actually, despite the name, the background colour is not the background colour.
    #background_colour_index = xf.background.background_colour_index
    #background_colour = book.colour_map[background_colour_index]

    return pattern_colour

production_color = getBGColor(wb,sh,0,0)
comprehension_color  = getBGColor(wb, sh, 0, 1)
```

```python
assert production_color == getBGColor(wb,sh,1,0)
```

```python
import tqdm
import xlrd

book = xlrd.open_workbook("English-Hebrew data for WordBank.xls",formatting_info=True)
list_of_dicts = {}
## {child1: {k1:"produce";
##          k2:"comprehend";
##          k3:""...};
##  child2: {...}
##  }

def check_type(book, sheet, row, column, values):
#     if sheet.name == "Template":
#         name = sheet.cell(row, column).value
#         if name:
#             columns.append(name)
#         return "skip"
    word = sheet.cell(row, column).value
    if type(word)!=float and word in values:
        word = word+"_2"
        
    color = getBGColor(book, sheet, row, column)
    if color == production_color:
        values[word] = "produce"
    elif color == comprehension_color:
        values[word] = "comprehend"
    else:
        if not sheet.cell_type(row, column) in (xlrd.XL_CELL_EMPTY, xlrd.XL_CELL_BLANK) or word!="":
            values[word] = ""
    return 
    

dataframe = []
## lop through all kids
for sheet in book.sheets():

#     new_row = [sheet.name]
    values = {}
    for row in range(sheet.nrows):
        ## if column B not empty: title column, skip
        if not sheet.cell_type(row, 1) in (xlrd.XL_CELL_EMPTY, xlrd.XL_CELL_BLANK):
            continue
        else:
            ## check for column C
            if sheet.cell(row, 2):
                t = check_type(book, sheet, row, 2, values)
                if t!="skip":
                    new_row.append(t)
                ## check for column D
                if sheet.cell(row, 3):
                    t = check_type(book, sheet, row, 3, values)
                    if t!="skip":
                        new_row.append(t)
                if sheet.cell(row, 4):
                    t = check_type(book, sheet, row, 4, values)
                    if t!="skip":
                        new_row.append(t)
                if sheet.cell(row, 5):
                    t = check_type(book, sheet, row, 5, values)
                    if t!="skip":
                        new_row.append(t)
    
    list_of_dicts[sheet.name] = values
    ## update dataframe
#     dataframe.append(new_row)


# c = ["CHILD"]
# c.extend(instruction["item"])
# output_df = pd.DataFrame(dataframe,columns=c)
# output_df
```
