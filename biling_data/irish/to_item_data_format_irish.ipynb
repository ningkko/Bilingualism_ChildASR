```python
import pandas as pd
import numpy as np
```

```python

irish = pd.read_csv("Irish_English.csv")
irish_data = pd.read_csv("Irish_English_data.csv")
```

```python
irish
```

```python
is_eng = []
for i,term in irish.iterrows():
    if term[5] == term[6]:
        is_eng.append(True)
    else:
        is_eng.append(False)
```

```python
irish["is_eng"] = is_eng
```

```python
english_terms = irish[irish["is_eng"] == True]
# english_terms
english_terms_id = english_terms["itemID"].to_list()
# english_terms_id
```

```python
irish_data
```

```python
keep_columns = list(irish_data.columns)
keep_columns = keep_columns[:10]+english_terms_id+keep_columns[-5:]
keep_columns
```

```python
english_in_irish = irish_data[keep_columns]
english_in_irish
```

```python
individual = english_in_irish.transpose()[:7]
individual.transpose().to_csv("individual_info.csv",index=False)
```
