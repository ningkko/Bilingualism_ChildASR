```python
import pandas as pd
import numpy as np
```

```python
instruction = pd.read_csv("wordbank/r01/Spanish_English_R01.csv").fillna("")
data = pd.read_csv("wordbank/r01/Spanish-English(R01, Bilingual).csv").fillna("")
instruction
```

```python
# data = data[:52]
data
```

```python
# instruction = instruction[instruction["lang"]=="eng"]
item_dict = dict(zip(instruction["itemID"],instruction["item"]))
item_dict
```

```python
columns = []
for c in list(data.columns):
    if c in item_dict:
        columns.append(item_dict[c])
    else:
        columns.append(c)
```

```python
data.columns = columns
columns = [c for c in columns if "item" not in c]
data = data[columns]
data
```

```python
def map_years(x):
    if x:
        return x*12
    return ""

data["age"] = data["age(years)"].apply(lambda x:map_years(x))
```

```python
data
```

```python
age = data.pop("age")
data.insert(1,"age",age)
data.pop("age(years)")
data
```

```python
data.to_csv("individual_info_R01.csv",index=False)
```

```python
## group the kids by eglish visit because we analyze only the english data
eg = data.groupby("age")
sessions = [eg.get_group(x) for x in eg.groups]
```

```python
data=data.transpose()
```

```python
d = {}
for session in sessions:
    childs = session["child"].to_list()
    column_name = "age_"+str(session["age"].unique()[0])
    df = session.transpose()[2:].replace("",np.nan)
    df["mean"] = df.mean(axis=1)
    df = df.astype(float).round(2)
#     if column_name=="age_34":
#         for i,j in df.iterrows():
#             if j["mean"]<0:
#                 print(i,j)

    d[column_name] = [childs,df["mean"]]
```

```python
d
```

```python
for k,v in d.items():
    data[k] = [str(v[0])]+[np.nan]+list(v[1])
```

```python
data
```

```python
data = data.transpose()[-3:].drop(columns=["age"]).transpose()
data
```
