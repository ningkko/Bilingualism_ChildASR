```python
import pandas as pd
import numpy as np
```

```python
irish = pd.read_csv("irish/Irish_English.csv")
irish_data = pd.read_csv("irish/Irish_English_data.csv")
```

```python
is_eng = []
for i,term in irish.iterrows():
    if term[5] == term[6]:
        is_eng.append(True)
    else:
        is_eng.append(False)
```

```python
irish["is_eng"] = is_eng
```

```python
english_terms = irish[irish["is_eng"] == True]
# english_terms
english_terms_id = english_terms["itemID"].to_list()
# english_terms_id
```

```python
keep_columns = list(irish_data.columns)
keep_columns = keep_columns[:10]+english_terms_id+keep_columns[-5:]
# keep_columns
```

```python
english_in_irish = irish_data[keep_columns]
```

```python
eg = english_in_irish.groupby("SessionNo")
sessions = [eg.get_group(x) for x in eg.groups]
```

```python
english_in_irish = english_in_irish.transpose()
```

```python
sessions[0]
```

```python
d = {}
for session in sessions:
    column_name = "session_"+str(session["SessionNo"].unique()[0])
    df = session.transpose()[10:-5]
    df["mean"] = df.mean(axis=1)
    d[column_name] = df["mean"]
```

```python
for k,v in d.items():
    english_in_irish[k] = [np.nan]*10+list(v)+[np.nan]*5 
```

```python
english_in_irish.index = keep_columns[:10]+english_terms["item"].to_list()+keep_columns[-5:]
```

```python
english_in_irish = english_in_irish.transpose()[-6:].transpose()
```

```python
english_in_irish = english_in_irish.astype(float).round(2)[10:]
```

```python
english_in_irish.to_csv("english_in_irish.csv")
```
