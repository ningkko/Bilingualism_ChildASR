```python
## for the instruction file, use one of the files
wave_raw = pd.read_excel("raw/MCDI Spreadsheet W1.xlsx", sheet_name="Wave 1")
wave_raw
```

#  wave 1 

```python
wave = wave_raw[1:62] ## the first 61 raws are subjects

subject_info = wave.transpose()[:25].transpose()  ## the forst 24 columns are subject information
data_comprehension_pruduction =  wave.transpose()[25:].transpose()

subject_columns = wave_raw.transpose()[0][:25]
subject_info.columns = subject_columns
subject_info
```

```python
data_comprehension_pruduction.transpose()
```

```python
new_data = []
for name,data in data_comprehension_pruduction.transpose().iterrows():
    new_row = []
    if "Unnamed" not in name:  ## this is the "understands" row
        for i in data:
            if i == 1:
                new_row.append("understands")
            elif i == 0:
                new_row.append("")
            else:
                new_row.append("no data entered")
    else:  ## "says"
        for i in data:
            if i == 1:
                new_row.append("produces")
            elif i == 0:
                new_row.append("")
            else:
                new_row.append("no data entered")
    new_data.append(new_row)
new_data_comprehension_pruduction = pd.DataFrame(new_data,index=data_comprehension_pruduction.columns)
data_comprehension_pruduction = new_data_comprehension_pruduction.transpose()
data_comprehension_pruduction
```

```python
data_comprehension_pruduction.transpose()
```

```python
new_data = []
first_row = []
for name,data in data_comprehension_pruduction.transpose().iterrows():
    combined_row = []
    if "Unnamed" in name: ## the second row in a word 
        for x,y in zip(first_row, data):
            if y and y!="no data entered":  ## says > understands, so if both "says" and "understands" == True, we record "says"
                combined_row.append(y)
            elif x:
                combined_row.append(x)
            else:
                combined_row.append("")
        new_data.append(combined_row)
    else: 
        first_row = data 

items = [i for i in list(wave_raw.columns) if "Unnamed" not in i]
new_data_comprehension_pruduction = pd.DataFrame(new_data,index=items)
data_comprehension_pruduction = new_data_comprehension_pruduction.transpose()
data_comprehension_pruduction
```

```python
subject_info = subject_info.transpose()
subject_info.columns = data_comprehension_pruduction.transpose().columns
data = pd.concat([subject_info, data_comprehension_pruduction.transpose()],axis=1)

data
```

```python

instruction = pd.DataFrame(items,columns=["item"])
first_french_index = instruction[instruction["item"]=="bêêê bêêê"].first_valid_index()
first_french_index
```

```python
instruction["lang"] = ["eng"]*396+["fr"]*(804-396)
instruction["choices"] = ["produces/undertands"]*len(instruction)
instruction["itemID"] = ["item_"+str(i) for i in list(range(1,len(instruction["item"])+1))]
instruction["complexity_category"] = ['']*len(instruction["item"])
# instruction["gloss"] = instruction["item"]
from collections import Counter
items = list(instruction.item)
uniq = []
for i in items:
    if i not in uniq:
        uniq.append(i)
    else:
        uniq.append(i+"_2")
instruction.item = uniq
```

```python
instruction.to_csv("wordbank/French_English_1.csv",index=False)
```
