```python
import pandas as pd
import numpy as np
```

```python
instruction = pd.read_csv("French_English_2.csv").fillna("")
data = pd.read_csv("french_english_data_2.csv").fillna("")
instruction
```

```python
# data = data[:52]
data
```

```python
def m(x):
    if not x:
        return ""
    if x == 1:
        return "English"
    elif x == 2:
        return "French"
    else:
        return "Other"

data["L1"] = data["L1"].apply(lambda x: m(x))
data["L2"] = data["L2"].apply(lambda x: m(x))
data["L3"] = data["L3"].apply(lambda x: m(x))


def m(x):
    if not x:
        return ""
    x = int(x)
    if x == 1:
        return "Male"
    elif x == 2:
        return "Female"
data["Sex"] = data["Sex"].apply(lambda x: m(x))


def m(x):
    if not x:
        return ""
    if x == 1:
        return "Monolingual"
    elif x == 2:
        return "Bilingual"
    
data["Language Grp"] = data["Language Grp"].apply(lambda x: m(x))

```

```python
instruction = instruction[instruction["lang"]=="eng"]
item_dict = dict(zip(instruction["itemID"],instruction["item"]))
item_dict
```

```python
columns = []
for c in list(data.columns):
    if c in item_dict:
        columns.append(item_dict[c])
    else:
        columns.append(c)
```

```python
data.columns = columns
columns = [c for c in columns if "item" not in c]
data = data[columns]
data
```

```python
list(data.columns)
```

```python
data = data.drop(columns = [ 'DOB','Wave', 'Maternal Education','Birth Order',
                             'MCDIp_E (# of words produced on MCDI)',
                             'MCDIpp_E (production percentile on MCDI',
                             'MCDIp_Fr (# of words produced on MCDI)',
                             'MCDIpp_Fr (production percentile on MCDI)',
                             'CDI Words'])
```

```python
data
```

```python
## Round up ages
def roundup(x):
    if not x:
        return ""
    return int(round(float(x),0))
data["Age at English Visit"] = data["Age at English Visit"].apply(lambda x: roundup(x))
data["Age at French visit"] = data["Age at French visit"].apply(lambda x: roundup(x))
data
```

```python
data.to_csv("individual_info(wave2).csv",index=False)
```

```python
## group the kids by eglish visit because we analyze only the english data
eg = data.groupby("Age at English Visit")
sessions = [eg.get_group(x) for x in eg.groups]
```

```python
data=data.transpose()
```

```python
d = {}
for session in sessions:
    childs = session["Subject"].to_list()
    column_name = "age_"+str(session["Age at English Visit"].unique()[0])
    df = session.transpose()[9:].replace("",np.nan)
    df["mean"] = df.mean(axis=1)
    df = df.astype(float).round(2)
#     if column_name=="age_34":
#         for i,j in df.iterrows():
#             if j["mean"]<0:
#                 print(i,j)

    d[column_name] = [childs,df["mean"]]
```

```python
d
```

```python
for k,v in d.items():
    data[k] = [str(v[0])]+[np.nan]*8+list(v[1])
```

```python
data = data.transpose()[-6:-1].transpose()[9:]
data
```
