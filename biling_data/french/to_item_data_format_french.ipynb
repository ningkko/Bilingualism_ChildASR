```python
import pandas as pd
import numpy as np
```

```python
instruction = pd.read_csv("wordbank/French_English_3.csv").fillna("")
data = pd.read_csv("wordbank/french_english_data_3.csv").fillna("")
instruction
```

```python
## wave 3
data = data[:52]

data
```

```python
def m(x):
    if not x:
        return ""
    if x == 1:
        return "English"
    elif x == 2:
        return "French"
    else:
        return "Other"

data["L1"] = data["L1"].apply(lambda x: m(x))
data["L2"] = data["L2"].apply(lambda x: m(x))
data["L3"] = data["L3"].apply(lambda x: m(x))


def m(x):
    if not x:
        return ""
    x = int(x)
    if x == 1:
        return "Male"
    elif x == 2:
        return "Female"
data["Sex"] = data["Sex"].apply(lambda x: m(x))


def m(x):
    if not x:
        return ""
    if x == 1:
        return "Monolingual"
    elif x == 2:
        return "Bilingual"
    
data["Language Grp"] = data["Language Grp"].apply(lambda x: m(x))

```

```python
instruction = instruction[instruction["lang"]=="eng"]
item_dict = dict(zip(instruction["itemID"],instruction["item"]))
item_dict
```

```python
print("\n".join(data.columns))
```

```python
columns = []
for c in list(data.columns):
    if c in item_dict:
        columns.append(item_dict[c])
    else:
        columns.append(c)
```

```python
data.columns = columns
columns = [c for c in columns if "item" not in c]
data = data[columns]
data
```

```python
## Round up ages
def roundup(x):
    if not x:
        return ""
    return int(round(float(x),0))
data["Age at English Visit"] = data["Age at English Visit"].apply(lambda x: roundup(x))
# data["Age at French visit"] = data["Age at French visit"].apply(lambda x: roundup(x))
data
```

```python
## group the kids by eglish visit because we analyze only the english data
eg = data.groupby("Age at English Visit")
sessions = [eg.get_group(x) for x in eg.groups]
```

```python
data=data.transpose()
```

```python
d = {}
for session in sessions:
    childs = session["Subject"].to_list()
    column_name = "age_"+str(session["Age at English Visit"].unique()[0])
    df = session.transpose()[2:].replace("",np.nan)
    df["mean"] = df.mean(axis=1)
    print(df)
    df = df.astype(float).round(2)
#     if column_name=="age_34":
#         for i,j in df.iterrows():
#             if j["mean"]<0:
#                 print(i,j)

    d[column_name] = [childs,df["mean"]]
```

```python
for k,v in d.items():
    data[k] = [str(v[0])]+[np.nan]*1+list(v[1])
data
```
